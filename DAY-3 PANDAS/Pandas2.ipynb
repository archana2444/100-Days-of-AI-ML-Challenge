## **Merging ,Joining & Concatenation**

```python
import numpy as np
import pandas  as pd
```

**Merging 2 dataframes**

```python
employees = pd.DataFrame({
    'employee_id':[101,102,103,104,105],
    'name':['Anjali','Yash','Raha','puja','mit'],
    'department':['HR','IT','Finance','IT','Marketing']
})

#dataframe2
salaries = pd.DataFrame({
    'employee_id':[101,102,103,7,8],
    'Salary':[60000,50000,40000,50000,45000],
    'Bonus':[2000,4000,6000,3000,4000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

### **Concatination of 2 Dataframes**

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
    })

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
    })
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

### **Joining 2 dataframes**

```python
df1 = pd.DataFrame({
    'name': ['Alaya', 'Bonny', 'Charles']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [76, 80, 98]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```

## **GroupBy Aggregation**

```python

data = {
    'Category': ['s1', 's2', 's1', 's2', 's1', 's2', 's1', 's2'],
    'Store': ['x', 'y', 'y', 'y', 'x', 'y', 'y', 'x'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

Group by categories and calculate the sum of sales

```python
col = df.groupby('Category')
col
for i,j in col:
    print(i)
    print(j)
```

```python
col = df.groupby('Category')['Sales'].sum()
col
```

Group by sales and calculate the sum of sales

```python
col = df.groupby('Store')['Sales'].sum()
col
```

**Group by multiple columns**
**Group by Category and Store**

```python
col = df.groupby(['Category','Store'])['Sales'].sum()
col
```

**Aggregation**

```python
df['Sales'].mean()
'''mean median mode min max count std'''
```

```python
df['Sales'].agg(['sum','mean','min','max','count','std','median'])
```

## **PIVOT TABLES**

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values="Sales",index="Region",columns="Product",aggfunc='median')
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

### **Cross Tabs**

```python
pd.crosstab(df['Region'],df['Product'])
```
