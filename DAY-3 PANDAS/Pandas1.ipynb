It is a one-dimensional array holding data of any type.

# **Pandas Dataframe**

### Creating Dataframe

```python
data = {
    'Name' : ['John','Aditi','Yash','puja'],
    'Age':[25,26,21,28],
    'City':['Delhi','Mumbai','Banglore','Pune'],
    'Salary':[65000,50000,45000,80000]
}
df= pd.DataFrame(data)
```

```python
df
```

## **Find missing data**

```python
data = {
    'A': [1,2,3,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

## **Removing missing Data**

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

## **Filling the missing Data**

```python
df
```

```python
df = df.fillna(0)
```

```python
df
```

```python
values = {'A':0,'B':1,"C": 20,'D':200}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
