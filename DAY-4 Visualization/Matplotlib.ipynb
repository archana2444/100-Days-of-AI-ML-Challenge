```python
import pandas as np
import numpy as np
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

Matplotlib is a library in Python that helps us draw graphs, charts, and plots to represent data visually.

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**3,x**2)
```

**Object-Oriented way**

Matplotlib allows the aspect ratio,DPI,and figure size to be specified when the figure object is created.You can use figsize and DPI Keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch(pixel per inch)

```python
fig = plt.figure(figsize=(6,8),dpi=100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```
