```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

1.   distplot/histplot

1.   Jointplot
2.   pairplot


2.   rugplot




```python
 df = sns.load_dataset('tips')
 df
```

```python
df['size'].unique()
```

```python
sns.histplot(df['total_bill'])
```

**Histograms** are visualization tools that represent the distribution of a set of continuous data. In a histogram, the data is divided into a set of intervals or bins (usually on the x-axis) and the count of data points that fall into each bin corresponding to the height of the bar above that bin. These bins may or may not be equal in width but are adjacent (with no gaps).  

```python
sns.distplot(df['total_bill'],bins=100)
```

 There are a variety of smoothing techniques. **Kernel Density Estimation (KDE)** is one of the techniques used to smooth a histogram.

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=30,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=10,kde=True)
```

# **Joint plot**
Syntax: seaborn.jointplot(x,  y,  data=None, kind='scatter', stat_func=None, color=None, height=6, ratio=5, space=0.2, dropna=True,  xlim=None, ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)


Parameters: The description of some main parameters are given below:


x, y: These parameters take Data or names of variables in "data".


data: (optional) This parameter take DataFrame when "x" and "y" are variable names.


kind: (optional) This parameter take Kind of plot to draw.


color:  (optional) This parameter take Color used for the plot elements.


dropna: (optional) This parameter take boolean value, If True, remove observations that are missing from "x" and "y".


Return: jointgrid object with the plot on it.



```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind = 'reg')
```

# **Pair plot**
seaborn.pairplot() method is used for visualizing relationships between multiple variables in a dataset. By creating a grid of scatter plots it helps to identify how different features interact with each other to identify patterns, correlations and trends in data. In this article, we will see how to implement seaborn.pairplot() in python.

We will use **hue parameter** to color-code points based on the day column. This helps to distinguish between different days of the week.

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```
