```python
import numpy as np
```

# **SECTION 1: Array Creation & Basics**
We'll simulate **365 days × 5 cities** of temperature data (in °C).

**Cities:** Mumbai, Delhi, Bangalore, Chennai, Kolkata

```python
# Temperature data: 7 days x 3 cities (rows = days, columns = cities)
# Each row is one day, each column is one city
temperatures = np.array([
    [32, 28, 24],   # Day 1
    [33, 30, 25],   # Day 2
    [31, 35, 23],   # Day 3
    [34, 36, 26],   # Day 4
    [30, 29, 22],   # Day 5
    [29, 27, 21],   # Day 6
    [35, 31, 25],   # Day 7
])

cities = ['Mumbai', 'Delhi', 'Bangalore']
days   = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

print('Shape of array:', temperatures.shape)  # (7 rows, 3 columns)
print('Total values  :', temperatures.size)   # 7 x 3 = 21
print('Data type     :', temperatures.dtype)  # int64
print('\nFull temperature table:')
print(temperatures)
```

```python
# Some basic array creation functions you should know

zeros   = np.zeros((3, 3))        # 3x3 array filled with 0
ones    = np.ones((2, 4))         # 2x4 array filled with 1
day_num = np.arange(1, 8)         # [1, 2, 3, 4, 5, 6, 7]

print('zeros array:\n', zeros)
print('ones array:\n', ones)
print('day numbers:', day_num)
```

**PART 2: Indexing & Slicing**

```python
# --- Get a single value ---
print('Day 1, Mumbai temp   :', temperatures[0, 0])   # row 0, col 0
print('Day 3, Delhi temp    :', temperatures[2, 1])   # row 2, col 1
print('Day 7, Bangalore temp:', temperatures[6, 2])   # row 6, col 2
```

```python
# --- Get an entire row or column ---

# All cities on Day 1 (row 0)
print('All cities on Day 1:', temperatures[0])

# All 7 days for Mumbai (column 0)
mumbai_temps = temperatures[:, 0]   # : means "all rows"
print('Mumbai all week    :', mumbai_temps)

# All 7 days for Delhi (column 1)
delhi_temps = temperatures[:, 1]
print('Delhi all week     :', delhi_temps)
```

```python
# --- Slice a range of days ---

# First 3 days only
first_3_days = temperatures[0:3]
print('First 3 days:\n', first_3_days)

# Last 2 days only
last_2_days = temperatures[-2:]
print('Last 2 days:\n', last_2_days)
```

```python
# --- Boolean Indexing (slightly tricky but very useful!) ---
# Find all temperatures above 33°C across all cities

hot_mask = temperatures > 33          # Creates a True/False array
print('Hot days mask (True = above 33°C):')
print(hot_mask)

print('\nActual hot temperatures:', temperatures[hot_mask])  # Only the True values
```

# **📊 PART 3: Math & Aggregate Functions**

NumPy makes it very easy to calculate stats — no loops needed!

```python
# --- Overall stats ---
print('Highest temp this week:', np.max(temperatures))
print('Lowest temp this week :', np.min(temperatures))
print('Average temp this week:', np.mean(temperatures))
```

```python
# --- Per-city stats (axis=0 means "across all rows/days") ---
city_avg = np.mean(temperatures, axis=0)  # One avg per city
city_max = np.max(temperatures, axis=0)
city_min = np.min(temperatures, axis=0)

print('Average temp per city:', city_avg)
print('Max temp per city    :', city_max)
print('Min temp per city    :', city_min)

# Which city had the highest average?
hottest_idx = np.argmax(city_avg)         # Returns the index of the max value
print(f'\nHottest city on average: {cities[hottest_idx]} ({city_avg[hottest_idx]:.1f}°C)')
```

```python
# --- Per-day stats (axis=1 means "across all columns/cities") ---
day_avg = np.mean(temperatures, axis=1)   # One avg per day

print('Average temp per day:')
for i, day in enumerate(days):
    print(f'  {day}: {day_avg[i]:.1f}°C')

hottest_day_idx = np.argmax(day_avg)
print(f'\nHottest day of the week: {days[hottest_day_idx]}')
```

---
## 🔀 PART 4: Reshaping & Broadcasting

### Two powerful NumPy features — let's keep it simple!

```python
# --- Reshape: change the shape without changing the data ---

flat = temperatures.flatten()       # Turn 2D -> 1D (all 21 values in a line)
print('Flattened shape:', flat.shape)
print('Flattened array:', flat)

back_to_2d = flat.reshape(7, 3)    # Turn 1D -> 2D again
print('\nRestored shape:', back_to_2d.shape)
```

```python
# --- Broadcasting: apply an operation across the whole array at once ---
# Example: Convert all temperatures from Celsius to Fahrenheit
# Formula: F = C * 9/5 + 32

temps_f = temperatures * 9/5 + 32   # This applies to ALL 21 values at once!

print('Temperatures in Celsius (original):')
print(temperatures)
print('\nTemperatures in Fahrenheit:')
print(np.round(temps_f, 1))
```

```python
# --- Slightly tricky broadcasting: subtract each city's average ---
# This tells us how much each day was ABOVE or BELOW the weekly average

city_avg = temperatures.mean(axis=0)          # Shape: (3,)  — one avg per city
difference = temperatures - city_avg          # Shape: (7,3) - (3,) -> NumPy handles this!

print('City averages:', city_avg)
print('\nDifference from average (+ = hotter than usual, - = cooler):')
print(difference)
print('\nFor example, Mumbai Day 7 was', difference[6, 0], '°C above its weekly average')
```

## **🏁 FINAL: Weekly Summary**

Put it all together — a simple report using everything above.

```python
print('=' * 40)
print('    📋 WEEKLY WEATHER SUMMARY')
print('=' * 40)

for i, city in enumerate(cities):
    col = temperatures[:, i]          # All 7 days for this city
    print(f'\n🏙️  {city}')
    print(f'   Average : {col.mean():.1f}°C')
    print(f'   Highest : {col.max()}°C  ({days[np.argmax(col)]})')
    print(f'   Lowest  : {col.min()}°C  ({days[np.argmin(col)]})')
    print(f'   Hot days (>33°C): {np.sum(col > 33)}')

print('\n' + '=' * 40)
print('Day 2 ✅ Complete! See you on Day 3! 🚀')
print('=' * 40)
```

## 🧠 Quick Recap

| What | How |
|---|---|
| Create array | `np.array(...)`, `np.zeros()`, `np.arange()` |
| Get one value | `arr[row, col]` |
| Get full column | `arr[:, col]` |
| Get full row | `arr[row]` |
| Filter values | `arr[arr > 33]` |
| Stats | `np.mean()`, `np.max()`, `np.min()` |
| Stats per column | `np.mean(arr, axis=0)` |
| Stats per row | `np.mean(arr, axis=1)` |
| Reshape | `arr.flatten()`, `arr.reshape(r, c)` |
| Convert all values | `arr * 9/5 + 32` (broadcasting) |
